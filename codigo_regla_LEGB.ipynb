## GLOBAL SCOPE



```python
x = 88 		# X es la variable global

def func():
	x = 99 	# El local oculta la global

func()
print(x) #	 Se imprime la original
```

```python
x = 88 			# Global X

def func():
	global x
	x = 99 		# La global x cambia a 99

func()
print("x =", x) # Imprime 99
```

## LEGB rule

```python


y, z = 1, 2         #Variables globales

def all_global():
	global x        # Los declara globales
	x = y + z       # Más no hace falta declarar
                   

all_global()
print('x = %d, y = %d, z = %d' % (x, y, z))
```

## NESTED SCOPES


```python

x = 99                      

def f1():
    x = 88
    def f2():               # Se cierra la definicion local
        print("x local = ", x)    # Referencia a definición anidada
    f2()                    

f1()                        # Imprime88
print("x global = ", x)     # X es 88 pero la global sigue siendo 99

# f2()  NameError: name 'f2' is not defined  No se puede invocar a f2() desde el modulo principal
```

## Factory Functions: Closures



```python
def f1():
	x = 88                 
	def f2():
		print("x =", x)    
	return f2 	           

action = f1()       
action() 			

```

```python
def maker(n):
	def action(x):		#Retorna la acción
		return x ** n 	
	return action

f = maker(2)            # pasa 2 al argumento 
f(3)                    # pasa 3 al x

f(4)		            # retorna 4 elevado a 2
```
