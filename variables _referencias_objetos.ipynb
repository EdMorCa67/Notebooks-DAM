## OBJETOS MUTABLES

### Listas

```python

L = [2, 3, 4]	    # Un objeto mutable
M = L			    # M obtiene mismos valores que L
L[0] = 24 		    # el valor de L cambia (el primer elemento de 24)

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4] , el valor tambi√©n cambia al ser mutable
```

Copia de objetos en vez de referencia:


```python

L = [2, 3, 4]
M = L[:]		    # Hace una copia de la lista L,  al tratarse de una copia esta vez la lista M no copia la L original
L[0] = 24           #La L original cambia de valores

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M no cambia
```

### Shared References and Equality
- Operador `==` igualdad de valores.
- Operador `is` identidad de objetos.

```python

L = [1, 2, 3]
M = L 			    # M y L referencian a los mismos valores

L == M 			    # Comprueba que tiene los mismos valores
L is M 			    # compara las referencias (los punteros)
```

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M y L referencia mismos valores pero son diferentes listas

print(
    "L == M",
    L == M 			# verdadero
)

print(
    "L is M", 	    # Falsos, son diferentes listas
    L is M
)
```

#### CACHE

Because small integers and strings are cached and reused, though, is tells us they reference the same single object.


```python

X = 42
Y = 42		   

X  == Y         # Si, mismos valores
X is Y 		    # Pero diferentes objetos
```
