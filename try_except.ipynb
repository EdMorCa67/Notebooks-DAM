```python
# File mergedexc.py (Python 3.X + 2.X)
sep = '-' * 45 + '\n'
print(sep + 'EXCEPTION RAISED AND CAUGHT')
try:
    x = 'spam'[99]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

# PRIMERA EXPLICACIÓN

En este código con el comando try intenta encontrar el carácter 99 en la palabra spam, al no existir se ejecuta el comando except que funciona para cuando no se dan estos casos en concreto dando lugar a imprimir except run, posteriormente a ello el comando finally se ejecuta siempre cuando el codigo anterior finaliza, y finalmente en la última línea de código al estar fuera se ejecuta de manera normal

```python
print(sep + 'NO EXCEPTION RAISED')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

# SEGUNDA EXPLICACIÓN
En el segundo caso pasa lo opuesto al primero, ya que esta vez si existe un tercer carácter en la palabra spam, y como consecuencia de ello no se ejecuta el comando except , por ende solo se ejecuta los dos últimos mensajes con el comando print

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
else:
    print('else run')
finally:
    print('finally run')
print('after run')
```

# TERCERA EXPLICACIÓN

Sucede algo parecido al caso anterior , solo que esta vez se le añade el else, que se ejecuta cuando no se activa el comando except reemplazando su mensaje por el del else: "else run"

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')
try:
    x = 1 / 0
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```
